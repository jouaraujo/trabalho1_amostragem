```python
import pandas as pd
from itertools import combinations
```

### Para renda

```python
df = pd.DataFrame(
    {
        "Domicilio" : [1, 2, 3, 4, 5, 6],
        "Renda" : [800, 4_200, 1_600, 500, 900, 2_000],
        "Numero_moradores" : [2, 4, 2, 2, 4, 1],
        "Numero_trabalhadores" : [2, 3, 1, 1, 2, 1]
    }
)
```

```python
#Renda
print(f'O total é {df.Renda.sum()}')
print(f'A média é {df.Renda.mean()}')
print(f'A variância é {df.Renda.var()}')
```

```python
#Número de moradores
print(f'O total é {df.Numero_moradores.sum()}')
print(f'A média é {df.Numero_moradores.mean()}')
print(f'A variância é {df.Numero_moradores.var()}')
```

```python
#Número de trabalhadores
print(f'O total é {df.Numero_trabalhadores.sum()}')
print(f'A média é {df.Numero_trabalhadores.mean()}')
print(f'A variância é {df.Numero_trabalhadores.var()}')
```

```python
domicilio = list(df.Domicilio)
total_combinacoes = 0

for i in combinations(domicilio, 2):
        print(i)
        total_combinacoes += 1

print(f'O total de combinações é {total_combinacoes}')
```
