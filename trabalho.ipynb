(d) o valor esperado  dado por:

$E\left(\hat{Y}_{ht}\right) = \hat{Y}_{ht(i)\cdot p(s)}$ \
p(s) = 1/15

```python
primeiro_momento = (s1*(1/5)) + (s2*(1/5)) + (s3*(1/5)) + (s4*(1/5)) + (s5*(1/5)) + (s6*(1/5)) + (s7*(1/5)) + (s8*(1/5)) + (s9*(1/5)) + \
                         (s10*(1/5)) + (s11*(1/5)) + (s12*(1/5)) + (s13*(1/5)) + (s14*(1/5)) + (s15*(1/5))

primeiro_momento
```

A varincia  dada por: 

$E\left(\hat{Y}_{ht}^{2}\right) - \left[E\left(\hat{Y}_{ht}\right)\right]^{2}$

```python
segundo_momento = (s1**2*(1/5)) + (s2**2*(1/5)) + (s3**2*(1/5)) + (s4**2*(1/5)) + (s5**2*(1/5)) + (s6**2*(1/5)) + (s7**2*(1/5)) + (s8**2*(1/5)) + (s9**2*(1/5)) + \
                         (s10**2*(1/5)) + (s11**2*(1/5)) + (s12**2*(1/5)) + (s13**2*(1/5)) + (s14**2*(1/5)) + (s15**2*(1/5))

segundo_momento
```

```python
var = segundo_momento - primeiro_momento**2
var
```
